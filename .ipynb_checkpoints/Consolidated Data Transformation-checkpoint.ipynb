------------

```python
# 01 get the unique user names from nust data #
unique_users_nust = pd.crosstab(data['Source User'],'frequency').reset_index()
```

```python
# 02 get the unique user names from names data downaload from internet  #
unique_users_new_names = pd.crosstab(names_dt['firstname'],'frequency').reset_index()
```

```python
len(unique_users_new_names), len(unique_users_nust)
```

```python
# 2.5 filling the gaps with Analia #
for i in range(3208,3768):
    unique_users_new_names.loc[i,'firstname'] = unique_users_new_names.loc[0,'firstname']
```

```python
# 03 create a data frame that includes the source users and new names #
unique_users_new_names['Source User'] = unique_users_nust['Source User']
unique_users_new_names
```

```python
# 04 set the indexes to sources user for nust data and the combine data #
unique_users_new_names.set_index(['Source User'], inplace=True)
data.set_index(['Source User'], inplace=True)
```

```python
# 04 continue .. #
data['New_name'] = unique_users_new_names['firstname']
data.reset_index(inplace=True)
```

```python
data.shape
```

```python
# 05 rename and drop the columns #
data.drop(['Source User'], axis=1, inplace=True)
print(data.shape)
data['Source User'] = data['New_name']
print(data.shape)
data.drop(['New_name'], axis=1, inplace=True)
print(data.shape)
#
```

```python
# 06 export the data #
data.to_csv('/home/arsal/Data/data.csv')
```
